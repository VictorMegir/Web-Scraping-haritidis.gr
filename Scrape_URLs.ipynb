```python
import requests
from bs4 import BeautifulSoup
```

```python
# Scrape all links
```

```python
all_links_file = open('links.txt', 'w+')
```

```python
def scrape_links(page_url: str) -> None:
    print('Scraping product URLs from page: ' + page_url)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    all_links_file = open('links.txt', 'a')
    
    all_products = soup.find_all('li', class_='item product-item')
    for product in all_products:
        link = product.div.div.div.a['href']
        all_links_file.write(link + '\n')
    
    all_links_file.close()
    
    next = soup.find('a', class_='next')
    if next is not None: scrape_links(next['href'])
    return
```

```python
starting_url = 'https://www.haritidis.gr/en/jewelry'
scrape_links(starting_url)
```
